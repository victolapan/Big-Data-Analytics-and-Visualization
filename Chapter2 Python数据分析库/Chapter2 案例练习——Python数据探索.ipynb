## 数据类型转换

```python
sales.dtypes
```

### 数值数据类型转换

```python
# 数据类型转换
sales.Quantity = sales.Quantity.astype('int')
sales.UnitPrice = sales.UnitPrice.astype('float')
```

```python
sales.dtypes
```

### 字符串类型转换

```python
sales.InvoiceNo = sales.InvoiceNo.astype('string')
sales.StockCode = sales.StockCode.astype('string')
sales.Description = sales.Description.astype('string')
sales.CustomerID = sales.CustomerID.astype('string')
sales.Country = sales.Country.astype('string')
```

```python
sales.dtypes
```

```python
# 字符串小写
sales.Description = sales.Description.str.lower()
```

```python
sales.head()
```

### 时间类型转换

```python
# 时间类型修改
sales.InvoiceDate = pd.to_datetime(sales.InvoiceDate)
```

```python
sales.dtypes
```

```python
# 从时间中提取月份
sales['Month'] = sales.InvoiceDate.dt.month
# 从时间中提取月份
sales['Day'] = sales.InvoiceDate.dt.day
# 从时间中提取星期
sales['Dayofweek'] = sales.InvoiceDate.dt.dayofweek
```

```python
sales.head()
```

### 列切片

### 条件索引

```python
# 筛选单价（UnitPrice）在1磅-10磅且销售量（Quantity）在1-10件的所有数据
condition = (sales['UnitPrice'] >= 1) & (sales['UnitPrice'] <= 10) & (sales['Quantity'] >= 1) & (sales['Quantity'] <= 10)
sales[condition]
```

## DataFrame运算

### 货币单位转换（标量运算）

```python
# 将UnitPrice列的英镑转换为人民币
exchange_rate = 9.06
sales.UnitPrice * exchange_rate
```

```python
(sales.UnitPrice * exchange_rate).to_frame()
```

### 计算小计（向量计算）

```python
sales.UnitPrice * sales.Quantity
```

### 格式化（函数运算）

```python
# 将结果保留三位小数
sales['UnitPrice'].map(lambda x: format(x, '.3f'))
```

```python
sales['UnitPrice'].map(lambda x: format(x, '.3f')).to_frame()
```

### 自定义函数计算

```python
# 将UnitPrice列的英镑转换为人民币
def pound_to_yuan(x, exchange_rate=9.06):
    return x * exchange_rate
```

```python
sales.UnitPrice.map(pound_to_yuan)
```

```python
sales.UnitPrice.map(pound_to_yuan).to_frame()
```
